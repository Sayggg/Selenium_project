```python
def scrape_tweet_data(link) -> dict():
  !apt update
  !apt install chromium-chromedriver
  !pip install selenium
  
  # set options to be headless, ..
  from selenium import webdriver
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  
  driver= webdriver.Chrome(options=options)
  #driver= webdriver.Chrome(options=options)
  from selenium import webdriver
  from selenium.webdriver.common.by import By
  from selenium.webdriver.common.keys import Keys
  from time import sleep
  driver.get(link)
  print(driver.page_source)
  sleep(3)
  initial_phrase="https://twitter.com/"
   # Python3 code to remove whitespace
  def remove(string):
    return string.replace(" ", "")
  time_stamp = driver.find_element(By.XPATH,".//time").text
  print(time_stamp)
  influencer=driver.find_element(By.XPATH,"(.//div[@class='css-1dbjc4n r-1awozwy r-18u37iz r-1wbh5a2 r-dnmrzs'])[3]").text
  influencer=influencer.lower()
  # Driver Program
  influencer=remove(influencer)
  influencer=initial_phrase+influencer
  print(influencer)
  promoter=driver.find_element(By.XPATH,".//div[@class='css-1dbjc4n r-1awozwy r-18u37iz r-1wbh5a2 r-dnmrzs']").text
  promoter=promoter.lower()
  #driver program
  promoter=remove(promoter)
  promoter=initial_phrase+promoter
  print(promoter)

  influencer_tweet=driver.find_element(By.XPATH,"(.//div[@data-testid='tweetText'])[1]").text
  #click=driver.find_element(By.XPATH,"//span[contains(text(),influencer_tweet)]"
  print(influencer_tweet)
  promotion_tweet=driver.find_element(By.XPATH,".//div[@data-testid]").text
  print(promotion_tweet)
  influencer_tweet_likes = driver.find_element(By.XPATH,".//div[@data-testid='like']").text
  print(influencer_tweet_likes)
  promoter_tweet_likes = driver.find_element(By.XPATH,".//div[@data-testid='like']").text
  print(promoter_tweet_likes)
  influencer_tweet_text=driver.find_element(By.XPATH,".//div[@data-testid='tweetText']").text
  print(influencer_tweet_text)
  promoter_tweet_text= driver.find_element(By.XPATH,"(.//div[@data-testid='tweetText'])[2]").text
  print(promoter_tweet_text)
  data={'time_stamp':[time_stamp],'influencer':[influencer],'promoter':[promoter],'influencer_tweet':[influencer_tweet],'promotion_tweet':[promotion_tweet],'influencer_tweet_likes':[influencer_tweet_likes],'promoter_tweet_likes':[promoter_tweet_likes],'influencer_tweet_text':[influencer_tweet_text],'promoter_tweet_text':[promoter_tweet_text]}
  print(data)
  import pandas as pd
  df_3=pd.DataFrame(data)
  df_3
  df1=pd.DataFrame()
  df1=pd.concat([df1,df_3])
  print(df1)
  #i=1
  return(df1)
```
